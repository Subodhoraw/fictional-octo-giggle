```python
import pandas as pd
import random

# 1. Define lists of sample data
titles = [
    "The Silent Echo", "Cyber Punk City", "Love in Paris", "The Last Samurai", "Galaxy Wars",
    "Deep Blue Sea", "Laugh Out Loud", "The Haunted Manor", "Speed Racer", "Historical Journey",
    "AI Revolution", "Forest Gump 2", "The Great Escape", "Comedy Central", "Dark Souls",
    "The Avenger", "Romantic sunset", "Space Odyssey", "Underwater World", "Ghost Busters 3",
    "Fast Cars", "Ancient Rome", "Robot takeover", "Funny Bones", "Scary Night",
    "Hero's Journey", "Wedding Bells", "Star Trekking", "Shark Attack", "The Clown",
    "Zombie Apocalypse", "Super Hero", "Kiss the Cook", "Mars Mission", "Dolphin Tale",
    "Vampire Diaries", "Nitro Boost", "King Arthur", "Tech World", "Jester's Court",
    "Mystery Mansion", "Caped Crusader", "First Date", "Alien Invasion", "Whale Watch",
    "Werewolf Howl", "Drift King", "Spartan", "Silicon Valley", "Stand Up Special"
]

genres = ["Action", "Sci-Fi", "Romance", "History", "Horror", "Comedy", "Drama", "Thriller"]
directors = ["Christopher Nolan", "Steven Spielberg", "Greta Gerwig", "James Cameron", "Quentin Tarantino", "Martin Scorsese"]

# 2. Generate 50 rows of random data
data = []
for i in range(50):
    row = {
        "Title": titles[i],
        "Genre": random.choice(genres),
        "Rating": round(random.uniform(3.0, 9.9), 1),  # Random rating between 3.0 and 9.9
        "Popularity": random.randint(50, 500),         # Random popularity score
        "Year": random.randint(1990, 2024),            # Random year
        "Director": random.choice(directors)
    }
    data.append(row)

# 3. Create DataFrame
df = pd.DataFrame(data)

# 4. Save to CSV
file_name = "movies.csv"
df.to_csv(file_name, index=False)

print(f"✅ Successfully created '{file_name}' with 50 rows.")
print("\nFirst 5 rows preview:")
print(df.head())
```
